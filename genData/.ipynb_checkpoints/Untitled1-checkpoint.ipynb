```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point
sns.set_style('whitegrid')
```

```python
file_path = "../../data/raw/indian_map/india-polygon.shp"
map_df = gpd.read_file(file_path) 
map_df_copy = gpd.read_file(file_path)
map_df.head()
```

```python
st_nm = ["Meghalaya", "Nagaland", "Mizoram", "Kerala", "Tamil Nadu", "Goa", "Manipur", "Andhra Pradesh", "Arunachal Pradesh", "Odisha", "Karnataka", "Tripura",\
        "Himachal Pradesh", "Jammu and Kashmir", "West Bengal", "Assam", "Delhi", "Madhya Pradesh", "Maharashtra", "Bihar", "Gujarat", "Punjab", "Uttar Pradesh",\
        "Rajasthan", "Haryana"]

st_idx = [16,21,22,23,24,24,26,26,28,28,29,29,30,31,31,31,31,32,32,33,33,33,34,36,36]
```

```python
st_df = pd.DataFrame({"st_nm": st_nm, "st_idx": st_idx})
# df.loc[len(df.index)] = ['Amy', 89, 93] 
st_df.loc[len(st_df.index)] = ["Uttarakhand", st_df.loc[st_df["st_nm"] == "Uttar Pradesh"]["st_idx"].values[0]]
st_df.loc[len(st_df.index)] = ["Jharkhand", st_df.loc[st_df["st_nm"] == "Bihar"]["st_idx"].values[0]]
st_df.loc[len(st_df.index)] = ["Chhattisgarh", st_df.loc[st_df["st_nm"] == "Madhya Pradesh"]["st_idx"].values[0]]
st_df.loc[len(st_df.index)] = ["Ladakh", st_df.loc[st_df["st_nm"] == "Jammu and Kashmir"]["st_idx"].values[0]]
st_df.loc[len(st_df.index)] = ["Telangana", st_df.loc[st_df["st_nm"] == "Andhra Pradesh"]["st_idx"].values[0]]
```

```python
st_df.tail()
```

```python
df_merged = pd.merge(left=map_df, right=st_df, on="st_nm", how="outer", indicator=True)
```

```python
df_merged = df_merged[df_merged["_merge"] == "both"]
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Patriarchy Index (Singh et al. (2021))')
df_merged.plot(column='st_idx',cmap='inferno',\
            linewidth=0.8, ax=ax, edgecolor='0', legend=True,\
            markersize=[39.739192, -104.990337])

for x, y, label in zip(map_df.geometry.centroid.x, map_df.geometry.centroid.y, map_df['st_nm']):
    ax.text(x, y, label, fontsize=8, ha='center', color='darkblue')

plt.show()
```
