```python

# Flinn and Heckman say that minimum accepted wage is a "stongly consistent" estimator of the truncation point. I do not know what 
# "strongly" consistent means but I'll keep it as such for now. 
# Given current data, a_m = a_f = 1.204

# Write a function to calculate log-likelihood given data, \mu and \sigma^2.

# Returns the negative of log likelihood so that when minimizing, I maximize the OG log likelihood.
def logL(x, a, data):
    # mu: mean, s: std dev, a: truncation point
    # For both males and females, the lowest accepted log wage is 1.204
    mu = x[0]
    s = x[1]
    norm_dist = stats.norm(loc=mu, scale=s) 
    F_a = norm_dist.cdf(a)
    ll = sum([np.log(norm_dist.pdf(d)/(1-F_a)) for d in data])
    return -ll
```
